```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.membership as mf
import matplotlib.pyplot as plt


## Değişkenlerin oluşturulması

waiting_time = np.arange(0, 71, 1)
distance = np.arange(0, 21, 1)
load_capacity = np.arange(0, 11, 1)
priority = np.arange(0, 101, 1)

## Üyelik fonksiyonlarının oluşturulması

waiting_time_low = mf.trimf(waiting_time, [0, 0, 25])
waiting_time_med = mf.trimf(waiting_time, [20, 30, 40])
waiting_time_high = mf.trimf(waiting_time, [35, 70, 70])

distance_low = mf.trimf(distance, [0, 0, 9])
distance_med = mf.trimf(distance, [7, 12, 15])
distance_high = mf.trimf(distance, [13, 20, 20])

load_capacity_low = mf.trimf(load_capacity,[0, 0 , 4])
load_capacity_med = mf.trimf(load_capacity,[2, 5 , 7])
load_capacity_high= mf.trimf(load_capacity,[6, 10 , 10])

priority_low = mf.trimf(priority,[0, 0 , 50])
priority_med = mf.trimf(priority,[25, 50 , 75])
priority_high= mf.trimf(priority,[60, 100, 100])

## Veri görselleştirme

fig, (ax0, ax1, ax2,ax3) = plt.subplots(nrows = 4, figsize =(6, 10))

ax0.plot(waiting_time, waiting_time_low, 'r', linewidth = 2, label = 'Düşük')
ax0.plot(waiting_time, waiting_time_med, 'g', linewidth = 2, label = 'Orta')
ax0.plot(waiting_time, waiting_time_high, 'b', linewidth = 2, label = 'Yüksek')
ax0.set_title('Waiting Time')
ax0.legend()

ax1.plot(distance, distance_low, 'r', linewidth = 2, label = 'Düşük')
ax1.plot(distance, distance_med, 'g', linewidth = 2, label = 'Orta')
ax1.plot(distance, distance_high, 'b', linewidth = 2, label = 'Yüksek')
ax1.set_title('Distance')
ax1.legend()

ax2.plot(load_capacity, load_capacity_low, 'r', linewidth = 2, label = 'Zayıf')
ax2.plot(load_capacity, load_capacity_med, 'g', linewidth = 2, label = 'Orta')
ax2.plot(load_capacity, load_capacity_high, 'b', linewidth = 2, label = 'Güçlü')
ax2.set_title('Load_Capacity')
ax2.legend()

ax3.plot(priority, priority_low, 'r', linewidth = 2, label = 'Zayıf')
ax3.plot(priority, priority_med, 'g', linewidth = 2, label = 'Orta')
ax3.plot(priority, priority_high, 'b', linewidth = 2, label = 'Güçlü')
ax3.set_title('Priority')
ax3.legend()

plt.tight_layout()

#Bulunduğunuz katı girdiğiniz alan
input_urfloor = 5

#Bulunduğunuz kattan bineceğiniz kişi sayısı
input_load_capacity = 3

#Asansörlerin bulunduğu kat bilgilerinin girilmesi
el1_floor = 1
el2_floor = 15
el3_floor = 17

#Asansör kapasitesi 10 kişi ile sınırlandırılmıştır.
elevator_limit = 10

#Asansörlerdeki kişi sayıları(yoğunluk)
el1_load=2
el2_load=6
el3_load=5

#Asansöre binen her kişi için bekleme süresi 3 saniye olarak belirlenmiştir.
getinside=3

el1_wtime=abs(input_urfloor-el1_floor)*2 + el1_load * getinside
el2_wtime=abs(input_urfloor-el2_floor)*2 + el2_load * getinside
el3_wtime=abs(input_urfloor-el3_floor)*2 + el3_load * getinside


# Üyelik derecelerinin hesaplanması

#elevator1
waiting_time_fit_low = fuzz.interp_membership(waiting_time, waiting_time_low, el1_wtime)
waiting_time_fit_med = fuzz.interp_membership(waiting_time, waiting_time_med, el1_wtime)
waiting_time_fit_hig = fuzz.interp_membership(waiting_time, waiting_time_high, el1_wtime)

load_capacity_fit_low = fuzz.interp_membership(load_capacity, load_capacity_low, el1_load)
load_capacity_fit_med = fuzz.interp_membership(load_capacity, load_capacity_med, el1_load)
load_capacity_fit_hig = fuzz.interp_membership(load_capacity, load_capacity_high, el1_load)

distance_fit_low = fuzz.interp_membership(distance, distance_low, abs(input_urfloor-el1_floor))
distance_fit_med = fuzz.interp_membership(distance, distance_med, abs(input_urfloor-el1_floor))
distance_fit_hig = fuzz.interp_membership(distance, distance_high, abs(input_urfloor-el1_floor))

#elevator2

waiting_time2_fit_low = fuzz.interp_membership(waiting_time, waiting_time_low, el2_wtime)
waiting_time2_fit_med = fuzz.interp_membership(waiting_time, waiting_time_med, el2_wtime)
waiting_time2_fit_hig = fuzz.interp_membership(waiting_time, waiting_time_high, el2_wtime)

load_capacity2_fit_low = fuzz.interp_membership(load_capacity, load_capacity_low, el2_load)
load_capacity2_fit_med = fuzz.interp_membership(load_capacity, load_capacity_med, el2_load)
load_capacity2_fit_hig = fuzz.interp_membership(load_capacity, load_capacity_high, el2_load)

distance2_fit_low = fuzz.interp_membership(distance, distance_low, abs(input_urfloor-el2_floor))
distance2_fit_med = fuzz.interp_membership(distance, distance_med, abs(input_urfloor-el2_floor))
distance2_fit_hig = fuzz.interp_membership(distance, distance_high, abs(input_urfloor-el2_floor))

#elevator3

waiting_time3_fit_low = fuzz.interp_membership(waiting_time, waiting_time_low, el3_wtime)
waiting_time3_fit_med = fuzz.interp_membership(waiting_time, waiting_time_med, el3_wtime)
waiting_time3_fit_hig = fuzz.interp_membership(waiting_time, waiting_time_high, el3_wtime)

load_capacity3_fit_low = fuzz.interp_membership(load_capacity, load_capacity_low, el3_load)
load_capacity3_fit_med = fuzz.interp_membership(load_capacity, load_capacity_med, el3_load)
load_capacity3_fit_hig = fuzz.interp_membership(load_capacity, load_capacity_high, el3_load)

distance3_fit_low = fuzz.interp_membership(distance, distance_low, abs(input_urfloor-el3_floor))
distance3_fit_med = fuzz.interp_membership(distance, distance_med, abs(input_urfloor-el3_floor))
distance3_fit_hig = fuzz.interp_membership(distance, distance_high, abs(input_urfloor-el3_floor))

# Kuralların oluşturulması

#el1

rule1=np.fmin(np.fmin(distance_fit_hig,waiting_time_fit_hig), priority_low)
rule2=np.fmin(np.fmin(distance_fit_hig,waiting_time_fit_low), priority_med)
rule3=np.fmin(np.fmin(distance_fit_hig,waiting_time_fit_med), priority_low)
rule4=np.fmin(np.fmin(distance_fit_low,waiting_time_fit_hig), priority_med)
rule5=np.fmin(np.fmin(distance_fit_low,waiting_time_fit_low), priority_high)
rule6=np.fmin(np.fmin(distance_fit_low,waiting_time_fit_med), priority_med)
rule7=np.fmin(np.fmin(distance_fit_med,waiting_time_fit_med),priority_med)
rule8=np.fmin(np.fmin(distance_fit_med,waiting_time_fit_hig),priority_low)
rule9=np.fmin(np.fmin(load_capacity_fit_low,waiting_time_fit_hig),priority_low)
rule10=np.fmin(np.fmin(distance_fit_low,load_capacity_fit_low), priority_low)
rule11=np.fmin(np.fmin(load_capacity_fit_med ,waiting_time_fit_low),priority_high)

#el2

rule12=np.fmin(np.fmin(distance2_fit_hig,waiting_time2_fit_hig), priority_low)
rule13=np.fmin(np.fmin(distance2_fit_hig,waiting_time2_fit_low), priority_med)
rule14=np.fmin(np.fmin(distance2_fit_hig,waiting_time2_fit_med), priority_low)
rule15=np.fmin(np.fmin(distance2_fit_low,waiting_time2_fit_hig), priority_med)
rule16=np.fmin(np.fmin(distance2_fit_low,waiting_time2_fit_low), priority_high)
rule17=np.fmin(np.fmin(distance2_fit_low,waiting_time2_fit_med), priority_med)
rule18=np.fmin(np.fmin(distance2_fit_med,waiting_time2_fit_med),priority_med)
rule19=np.fmin(np.fmin(distance2_fit_med,waiting_time2_fit_hig),priority_low)
rule20=np.fmin(np.fmin(load_capacity2_fit_low,waiting_time2_fit_hig),priority_low)
rule21=np.fmin(np.fmin(distance2_fit_low,load_capacity2_fit_low), priority_low)
rule22=np.fmin(np.fmin(load_capacity2_fit_med ,waiting_time2_fit_low),priority_high)

#el3

rule23=np.fmin(np.fmin(distance3_fit_hig,waiting_time3_fit_hig), priority_low)
rule24=np.fmin(np.fmin(distance3_fit_hig,waiting_time3_fit_low), priority_med)
rule25=np.fmin(np.fmin(distance3_fit_hig,waiting_time3_fit_med), priority_low)
rule26=np.fmin(np.fmin(distance3_fit_low,waiting_time3_fit_hig), priority_med)
rule27=np.fmin(np.fmin(distance3_fit_low,waiting_time3_fit_low), priority_high)
rule28=np.fmin(np.fmin(distance3_fit_low,waiting_time3_fit_med), priority_med)
rule29=np.fmin(np.fmin(distance3_fit_med,waiting_time3_fit_med),priority_med)
rule30=np.fmin(np.fmin(distance3_fit_med,waiting_time3_fit_hig),priority_low)
rule31=np.fmin(np.fmin(load_capacity3_fit_low,waiting_time3_fit_hig),priority_low)
rule32=np.fmin(np.fmin(distance3_fit_low,load_capacity3_fit_low), priority_low)
rule33=np.fmin(np.fmin(load_capacity3_fit_med ,waiting_time3_fit_low),priority_high)

# Birleşim kümelerinin oluşturulması


out_low = np.maximum.reduce([rule1,rule3,rule8,rule9,rule10])
out_med=np.maximum.reduce([rule2,rule4,rule6,rule7])
out_high=np.maximum.reduce([rule5,rule11])

out_low2= np.maximum.reduce([rule12,rule14,rule19,rule20,rule21])
out_med2=np.maximum.reduce([rule13,rule15,rule17,rule18])
out_high2=np.maximum.reduce([rule16,rule22])

out_low3= np.maximum.reduce([rule23,rule25,rule30,rule31,rule32])
out_med3=np.maximum.reduce([rule24,rule26,rule28,rule29])
out_high3=np.maximum.reduce([rule27,rule33])

# Durulaştırma

out_brake = np.maximum.reduce([out_low, out_med, out_high])
defuzzified  = fuzz.defuzz(priority, out_brake, 'centroid')
result = fuzz.interp_membership(priority, out_brake, defuzzified)

out_brake2 = np.maximum.reduce([out_low2, out_med2, out_high2])
defuzzified2 = fuzz.defuzz(priority, out_brake2, 'centroid')
result2 = fuzz.interp_membership(priority, out_brake2, defuzzified2)

out_brake3 = np.maximum.reduce([out_low3, out_med3, out_high3])
defuzzified3  = fuzz.defuzz(priority, out_brake3, 'centroid')
result3 = fuzz.interp_membership(priority, out_brake3, defuzzified3)

# Sonuç-1

print("(1-Öncelik) Çıkış Değeri:", defuzzified)
print("(1-Doluluk) Doluluk Değeri:",el1_load)
print("(2-Öncelik) Çıkış Değeri:", defuzzified2)
print("(2-Doluluk) Doluluk Değeri:",el2_load)
print("(3-Öncelik) Çıkış Değeri:", defuzzified3)
print("(3-Doluluk) Doluluk Değeri:",el3_load)
print("")

#Kapasite kontrolleri
control1=0
control2=0
control3=0

if el1_load+input_load_capacity >= elevator_limit:
    control1 = 1
    defuzzified = 0
    print("Asansör 1 tam kapasitede çalışmaktadır.")
    print("")
    
if el2_load+input_load_capacity >= elevator_limit:
    control2 = 1
    defuzzified2 = 0
    print("Asansör 2 tam kapasitede çalışmaktadır.")
    print("")
    
if el3_load+input_load_capacity >= elevator_limit:
    control3 = 1
    defuzzified3 = 0
    print("Asansör 3 tam kapasitede çalışmaktadır.")
    print("")
    
    
#Bulunan katta asansör varlığı kontrolü
    
if input_urfloor==el1_floor and control1==0 :
    print("Asansör 1 bulunduğunuz kattadır.")
    
if input_urfloor==el2_floor and control2==0:
    print("Asansör 2 bulunduğunuz kattadır.")
    
if input_urfloor==el3_floor and control3==0 :
    print("Asansör 3 bulunduğunuz kattadır.")    
    
#Sonuç-2
sfta = max(defuzzified,defuzzified2,defuzzified3)

if sfta == defuzzified and control1 != 1  and input_urfloor!=el1_floor:
    print("(Öncelik) Çıkış Değeri:", defuzzified)
    print("Asansör 1 yönlendiriliyor ...")

elif sfta == defuzzified2 and control2 != 1 and input_urfloor!=el2_floor :
    print("(Öncelik) Çıkış Değeri:", defuzzified2)
    print("Asansör 2 yönlendiriliyor ...")

elif sfta == defuzzified3 and control3 != 1 and input_urfloor==el3_floor:
    print("(Öncelik) Çıkış Değeri:", defuzzified3)
    print("Asansör 3 yönlendiriliyor ...")

    
print("")
print("Girdilerin grafikleri:")    
#pip install -U scikit-fuzzy kurulmalıdır.

```
